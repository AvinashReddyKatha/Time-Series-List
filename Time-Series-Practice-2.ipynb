```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("aapl_no_dates.csv")
df.tail()
```

```python
rng=pd.date_range(start="8/1/2024",end="8/30/2024",freq='B')
rng
```

```python
df.set_index(rng,inplace=True)
df
```

```python
df.asfreq('D',method='pad')

#This fills the weekend data based on previous days
```

```python
df['Open'].plot(kind='line', title='Open Prices')
plt.xlabel('Date')
plt.ylabel('Open Price')
plt.show()
```
